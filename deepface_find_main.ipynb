```python
from face_authenticator import FaceAuthenticator
import matplotlib.pyplot as plt
import os
```

```python
authenticator = FaceAuthenticator(db_path='db')
authenticator.plot_faces()
```

```python
test_folder = 'test'
scores = []
real = []

for image in tqdm(os.listdir(test_folder)):
    img_path = os.path.join(test_folder, image)
    scores.append(authenticator.predict(img_path, plot=False))
    real.append(1 if 'rafa' in image else 0)
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(real, scores)*100
print('test accuracy : {:.2f}%'.format(acc))
```
